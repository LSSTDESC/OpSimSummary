```python
synopsim = SynOpSim.fromOpSimDB(myopsimv3, opsimversion='lsstv3',
                                angleUnit='radians', usePointingTree=True)
```

`SynOpSim` accesses the pointings that will observe a particular point (ra, dec) for a sequence of (ra, dec). Since a pointing contains lots of data, and we don't want to put everything in memory, we use a python generator. The syntax  is

gen = pointingsEnclosing(raSeq, decSeq, pointingRadius=1.75, usePointingTree=True)

Then we can iterate along the elements of raSeq, decSeq using next(gen) which gives a stop iteration error when all of the elements are exhausted.

```python
gen = synopsim.pointingsEnclosing([54.0, 52.], [-27.5, -23.0] , circRadius=0., pointingRadius=1.75, 
                                  usePointingTree=True)
```
